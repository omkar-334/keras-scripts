```python
import numpy as np
import torch
import tensorflow as tf
from transformers import AutoModel, AutoTokenizer
```

 # Load HF Model and Tokenizer

```python
from transformers import RobertaTokenizer, TFRobertaModel
import tensorflow as tf

hf_model = TFRobertaModel.from_pretrained("roberta-base")
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
```

```python
text = "Hello, how are you?"

inputs = tokenizer(text, return_tensors="tf", padding=True, truncation=True)

hf_output = hf_model(**inputs).last_hidden_state
```

# Load Keras Model

```python
import keras_hub
model = keras_hub.models.RobertaBackbone.from_preset("hf://FacebookAI/roberta-base", dtype="bfloat16")
```

```python
model.summary()
```

```python
keras_inputs = {
    "token_ids": inputs["input_ids"].numpy(),  # Token IDs
    "padding_mask": inputs["attention_mask"].numpy(),  # Padding Mask
}

keras_output = model(keras_inputs)
```

# Check shapes and outputs

```python
print(hf_output.shape)
print(keras_output.shape)
```

```python
print("Huggingface output\n", hf_output)
print("\nKeras output\n", keras_output)
```

```python
import numpy as np

np.testing.assert_allclose(
    tf.cast(keras_output, dtype=tf.float32),
    tf.cast(hf_output, dtype=tf.float32),
    atol=1e-5
    )
```

# Load Keras RoBERTa Tokenizer

```python
import subprocess

def find_file(filename):
    try:
        path = subprocess.check_output(f'find /root/.cache/huggingface -name {filename}', shell=True, text=True).strip()
        if path:
            return path.split("\n")[0]
        else:
            raise FileNotFoundError(f"{filename} not found!")
    except subprocess.CalledProcessError:
        raise FileNotFoundError(f"Error finding {filename}!")
```

```python
from transformers import RobertaTokenizer
from keras_hub.tokenizers import RobertaTokenizer as KerasRobertaTokenizer


vocab_path = find_file("vocab.json")
merges_path = find_file("merges.txt")

keras_tokenizer = KerasRobertaTokenizer(vocabulary=vocab_path, merges=merges_path)
```

# Test tokenizer outputs

```python
text = "RoBERTa is a transformers model pretrained on a large corpus."

hf_tokens = hf_tokenizer.encode(text)
keras_tokens = keras_tokenizer.tokenize(text)

print("Hugging Face Tokenizer Output:\n", hf_tokens)
print("\nKeras Tokenizer Output:\n", keras_tokens)
```
